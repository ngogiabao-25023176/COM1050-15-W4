```python
#W4A2:
check=False
while not (check):
    n=int(input())
    if n>0:
        check=True
if n < 2:
    print(n, "không phải là số nguyên tố")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(n, "không phải là số nguyên tố")
            break
    else:
        print(n, "là số nguyên tố")
```
