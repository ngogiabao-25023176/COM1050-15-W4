# A10

```python
def collatz(n):
    dem=0
    while n!=1:
        if n%2==0:
            n//=2
        else:
            n=n*3+1
        dem+=1
    return dem

n = int(input())
lst=[]
res=[]

for i in range (1,n+1):
    lst.append(collatz(i))

for a in range(len(lst)):
    if lst[a]==max(lst):
        res.append(a+1)

print(max(lst),min(res))
```
